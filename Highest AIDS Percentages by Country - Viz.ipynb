```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Pulled HIV indicator from Gapminder through Github Repository. 

df=pd.read_csv('adults_with_hiv_percent_age_15_49.csv',index_col='country')
print(df.shape)
print(df.columns)
hiv_data = df.T
```

```python
#Calculated average percentages of adults infected with HIV Aids for each country. 
#Identifed the top 6 countries with highest average percentages.
#Created a scatter plot; trend seems linear.

df['average']=df.mean(axis=1)
top_6 = df.nlargest(6, 'average')
plt.scatter(x=top_6.index,y=top_6['average'],color='purple', marker = 'x')
plt.title('Average Percentage of Adult HIV (Top 6 Countries)')
plt.xlabel('Country')
plt.ylabel('% of Adult Population')
plt.show()
```

```python
#Aggregated all percentages across years for the top 6 countries with highest percentages of adults with HIV.
#Distribution is bimodal and skewed to the left.

plt.figure(figsize=(10,5))
top6_values=hiv_data['Botswana'].tolist()+hiv_data['Zimbabwe'].tolist()+hiv_data['Lesotho'].tolist()+hiv_data['Swaziland'].tolist()+hiv_data['Zambia'].tolist()+hiv_data['South Africa'].tolist()
plt.hist(top6_values, color='red',bins=40,alpha=0.4)
plt.title('Percentages of Adults with HIV (Top 6 Countries)')
```

```python
hiv_data=hiv_data.iloc[11:33]
hiv_data.index = pd.to_numeric(hiv_data.index)
```

```python
#Plotted the 6 countries with highest percentages of HIV over time. 
#Percentages converge for most countries.

plt.figure(figsize=(20,5))
plt.plot(hiv_data['Botswana'],color='red')
plt.plot(hiv_data['Zimbabwe'],color='green')
plt.plot(hiv_data['Lesotho'],color='blue')
plt.plot(hiv_data['Swaziland'],color='purple')
plt.plot(hiv_data['Zambia'],color='orange')
plt.plot(hiv_data['South Africa'],color='black')
plt.title('Percent of Adults with HIV over Time (Top 6 Countries)')
plt.xlabel('Year')
plt.ylabel('% of Adult Population')
plt.legend()
plt.show()
```

```python
# Compared United States and Botswana, considering Botswana had the highest average percentage of population with HIV.
# United States data is not continuous, discrete perecetange points, modulating from 0.5 to 0.6%. 

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(hiv_data['United States'], color = 'red')
plt.ylabel('% of Adult Population')
plt.title('Percent of Adults with HIV in United States')

plt.subplot(1,2,2)
plt.plot(hiv_data['Swaziland'], color = 'green')
plt.title('Percent of Adults with HIV in Botswana')
plt.tight_layout()
plt.show()
```
